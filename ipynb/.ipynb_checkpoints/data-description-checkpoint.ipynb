### Split Dataset to Training set & Test Set

```r
install.packages('caTools')
```

```r
install.packages('MASS')
```

```r
library(MASS)
library(caTools)
```

```r
set.seed(123)
split = sample.split(data_abalone$rings, SplitRatio = 0.7)
training_set = subset(data_abalone, split == TRUE)
test_set = subset(data_abalone, split == FALSE)
```

```r
dim(training_set)
```

### Feature Scaling

This is not applicable for Liner Regression. This is automatically taken care by function we will use.

### Fitting The Multiple Linear Regressor to Training Set

```r
regressor = lm(formula = rings ~ .,
             data = training_set)
```

```r
summary(regressor)
```

### Predicting the Test set results

```r
y_pred = predict(regressor, newdata = test_set)
```

```r
y_pred
```

### Backward Elimination

```r
regressor = lm(formula = rings ~ sex + length + diameter + height + whole_weight + shucked_weight + viscera_weight + shell_weight,
             data = data_abalone)
summary(regressor)
```

**Remove Length, since its significance is low or no impact**

```r
regressor = lm(formula = rings ~ diameter + height + whole_weight + shucked_weight + viscera_weight + shell_weight,
             data = data_abalone)
summary(regressor)
```
