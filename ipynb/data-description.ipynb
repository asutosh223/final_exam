```r
# abalone proportions
p_infants <- sum(abalone$SEX == "I")/length(abalone$SEX)
p_females <- sum(abalone$SEX == "F")/length(abalone$SEX)
p_males <- sum(abalone$SEX == "M")/length(abalone$SEX)
```

```r
abalone_mydata <- abalone[sample(1:4141, 500),]
write.table(abalone_mydata, file = "/home/jovyan/final_exam/data/abalone_data.csv")
```

```r
prop.test(sum(abalone_mydata$SEX == "I"), nrow(abalone_mydata), p = p_infants, conf.level = 0.95, correct = FALSE)
prop.test(sum(abalone_mydata$SEX == "F"), nrow(abalone_mydata), p = p_females, conf.level = 0.95, correct = FALSE)
prop.test(sum(abalone_mydata$SEX == "M"), nrow(abalone_mydata), p = p_males, conf.level = 0.95, correct = FALSE)
```
