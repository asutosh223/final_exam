
### Linear Regression Analysis

To perform the analysis, we will consider Rings as Dependpant Variable DV & perform Multiple Linear Regression.

In below equation y = Rings

y = b0 + b1*x1 + b2*x2 + .... + bn * xn

Assumption of Liner Regression
 - Linearity
 - Homoscedasticity
 - Multivariant normality
 - Independence of errors
 - Lack of multicollinearity
 
Sex is categorical variable.

```r
install.packages("rms")
library(MASS)
library(rms)
```

```r
abfit1 = lm(Rings ~ Sex + Length + Diameter + Height + Whole_weight 
               + Shucked_weight + Viscera_weight + Shell_weight, data = abalone)
abfit2 = stepAIC(abfit1)
summary(abfit2)
```

The sex being used as the baseline in this case is female. Since male is not significantly different, and infant is the only difference, we can redefine this feature to define whether gender has been expressed (infant vs non-infant)



```r
abalone$Sex = as.character(abalone$Sex)
abalone$Sex[abalone$Sex != 'I'] = 'K'
abalone$Sex = as.factor(abalone$Sex)
```

Whole Weight = Shucked Weight + Viscera Weight + Shell Weight + Unknown mass of water/blood lost from shucking process

We should investigate this abalone weight problem, and ensure that the variables in the model are not functions of eachother.

```r
abalone$weight.diff = abalone$Whole_weight - 
      (abalone$Viscera_weight + abalone$Shucked_weight + abalone$Shell_weight)
par(mfrow=c(1,1))
hist(abalone$weight.diff,breaks=50)
```

```r
abalone[abalone$weight.diff < -.1,]
```

```r
abfit2.1 = lm(Rings ~ Sex + Diameter + Height + Whole_weight, data = abalone)
abfit2.2 = lm(Rings ~ Sex + Diameter + Height + Shucked_weight, data = abalone) # !
abfit2.3 = lm(Rings ~ Sex + Diameter + Height + Viscera_weight, data = abalone)
abfit2.4 = lm(Rings ~ Sex + Diameter + Height + Shell_weight, data = abalone)
```

```r
abalone$weight.mean1 = (abalone$Shucked_weight*abalone$Viscera_weight*abalone$Shell_weight)^(1/3)
abalone$weight.mean2 = (abalone$Whole_weight*abalone$Shucked_weight*abalone$Shell_weight*abalone$Viscera_weight)^(1/4)
abalone$weight.norm1 = sqrt(abalone$Shucked_weight^2 + abalone$Viscera_weight^2 + abalone$Shell_weight^2)
abalone$weight.norm2 = sqrt(abalone$Whole_weight^2 + abalone$Shucked_weight^2 + abalone$Viscera_weight^2 + abalone$Shell_weight^2)
```

Well also define some other measures of size that incorporate all 3 dimensions. This allows us to describe 3 dimensions with 1 (hopefully). We pick the euclidean norm of its size, along with the geometric mean of size. Well pick one of these measures to use as well.

```r
abalone$size.norm = sqrt(abalone$Length^2 + abalone$Diameter^2 + abalone$Height^2) # Norm of vectors
abalone$size.mean = (abalone$Length*abalone$Diameter*abalone$Height)^(1/3)         # Geometric Mean
```

```r
abfit3 = lm(Rings ~ Sex + size.norm + Whole_weight + Shucked_weight + Viscera_weight + Shucked_weight, data = abalone)
 
abfit3.1 = lm(Rings ~ Sex + size.norm + Whole_weight, data = abalone)
abfit3.2 = lm(Rings ~ Sex + size.norm + Shucked_weight, data = abalone)
abfit3.3 = lm(Rings ~ Sex + size.norm + Viscera_weight, data = abalone)
abfit3.4 = lm(Rings ~ Sex + size.norm + Shucked_weight, data = abalone) # best
 
abfit4.1 = lm(Rings ~ Sex + size.norm + weight.norm1, data = abalone)
abfit4.2 = lm(Rings ~ Sex + size.norm + weight.norm2, data = abalone)
abfit4.3 = lm(Rings ~ Sex + size.norm + weight.mean1, data = abalone)
abfit4.4 = lm(Rings ~ Sex + size.norm + weight.mean2, data = abalone)
 
abfit5.1 = lm(Rings ~ Sex + size.mean + weight.norm1, data = abalone)
abfit5.2 = lm(Rings ~ Sex + size.mean + weight.norm2, data = abalone)
abfit5.3 = lm(Rings ~ Sex + size.mean + weight.mean1, data = abalone)
abfit5.4 = lm(Rings ~ Sex + size.mean + weight.mean2, data = abalone)
 
abfit6 = lm(Rings ~ Sex + size.mean + Whole_weight + Shucked_weight + Viscera_weight + Shucked_weight, data = abalone)
abfit6.1 = lm(Rings ~ Sex + size.mean + Shucked_weight + Viscera_weight + Shucked_weight, data = abalone)
abfit6.2 = lm(Rings ~ Sex + size.mean + Shucked_weight + Shucked_weight, data = abalone)
abfit6.3 = lm(Rings ~ Sex + size.mean + I(size.mean^2) + Shucked_weight + Shucked_weight, data = abalone)
```

```r
par(mfrow=c(2,2))
plot(abfit6.3)
```

When we look at the residuals vs fitted values plot, we see a fan shape indicating that as the fitted values increase, so to do the scale of the residuals. One way to account for that is to use a log transformation of the dependent variable. Subsequent analysis will take log(rings) as the dependent variable to account for this.



```r
abfit7 = lm(log(Rings) ~ Sex + size.mean + Shucked_weight + Shell_weight, data = abalone)
plot(abfit7)
```
